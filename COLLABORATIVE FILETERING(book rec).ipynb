# Popularity based

# now merge book with rating on ISBN
# then check each Book rating

```python
book_rating=book.merge(rating, on = 'ISBN')
book_rating.head(5)
```

```python
num_of_rating=book_rating.groupby('Book-Title').count()['Book-Rating'].reset_index() #each Book has total rating
num_of_rating=num_of_rating.rename(columns={'Book-Rating':'num_of_rating'}) #ranme Book-Rating to num_of_rating
num_of_rating
```

```python
#calulate mean of each book
avg_rating=book_rating.groupby('Book-Title').mean()['Book-Rating'].reset_index() #each Book has average rating
avg_rating=avg_rating.rename(columns={'Book-Rating':'avgrating'}) #ranme Book-Rating to avg_rating
avg_rating
```

```python
# now sort on the base of average rating
# num_of_rating and avg_rating merge
popular_df=num_of_rating.merge(avg_rating,on='Book-Title')
popular_df
```

```python
# books has rating >=250 and top 50
popular_df=popular_df[popular_df['num_of_rating']>=250].sort_values('avgrating',ascending=False).head(50)
popular_df.head(4)
```

```python
popular_df=popular_df.merge(book,on='Book-Title')
popular_df.head(1)
```

```python
# popular_df.duplicated('Book-Title').sum(),
popular_df=popular_df.drop_duplicates('Book-Title')[['Book-Title','num_of_rating','avgrating','Book-Author','Image-URL-M']]
popular_df.head(4)
```

# Colaborative Filetering
## now we need Users column and Books index  and their values Rating

```python
book_rating.head(1)
```

```python
book_rating.shape
```

```python
# get users who rated book > 200
book_rating.groupby('User-ID').count()['Book-Rating']>200
```

```python
x=book_rating.groupby('User-ID').count()['Book-Rating']>200
readr_users=x[x].index #get those users index(user_id) whose rating >200
readr_users.shape
```

```python
# now get those readers in BOOK_Rating DF
filtered_rating=book_rating[book_rating['User-ID'].isin(readr_users)]
filtered_rating.shape
```

## those user have >200 rate book 
#### filtered_rating
## and single book has >50 rating

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>50
famous_book=y[y].index
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
final_rating.shape
```

```python
# pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
pt
```

## get distance of BOOK 1 to Other BOOKs (similarity Score)

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores.shape #that is 
```

```python
# similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
# similarity_scores[1] enumerate to get index
list(enumerate(similarity_scores[5]))[1:5] # got sorted list with similarity but not sorted
```

```python
# filter having high similarity  on sort
sorted(list(enumerate(similarity_scores[5])),key=lambda x:x[1],reverse=True)[1:6]
```

```python
def recommend(book_name):
    p=pt.index==book_name
    if p[p].size>0:
        print("not")
        index=np.where(pt.index==book_name)[0][0]
        similar_book= sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
        data=[]
        for i in similar_book:
            items=[]
    #         item=i[0]
    #         print(pt.index[i[0]])
            df_temp=final_rating[final_rating['Book-Title']==pt.index[i[0]]]
    #         print(df_temp.drop_duplicates('Book-Title')['Book-Title'].values)
            items.extend(list(df_temp.drop_duplicates('Book-Title')['Book-Title'].values))
            items.extend(list(df_temp.drop_duplicates('Book-Title')['Book-Author'].values))
            items.extend(list(df_temp.drop_duplicates('Book-Title')['Image-URL-M'].values))

            data.append(items)
    
    else:
        
        print("empty")
        return False
        

    
```
